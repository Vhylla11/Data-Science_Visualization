```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
usia =[22,24,25,26,27,28,34,30,33,53,52,54,20,231]
kelompok =[20,25,30,35,40,45,50,55]
plt.hist(usia,kelompok,histtype='stepfilled',rwidth=10)
plt.xticks(np.arange(20,60,5))
plt.show()
```
