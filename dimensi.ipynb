# Korelasi

```python
df = pd.read_csv('atlit.csv')
df.corr()

## lihat korelasi dari gambar
plt.imshow(df.corr())
plt.colorbar()
col = list(map(lambda x:x[1],df.columns.tolist()))
i = list(map(lambda x:x,df.index.tolist()))
print(i)

for x in range (len(i)) :
    for y in range (len(col)):
        plt.text(y,x, df.iloc[x,y],color='r',fontsize=20,ha='center',va='center')
plt.set_xticks(np.arange(len(col)))
plt.set_xticklabels(col)
plt.set_yticks(np.arange(len(i)))
plt.set_yticklabels(i)
plt.set_ylim(2.5,-0.5)
plt.show()
plt.show()
```

```python
## buat 2 dimensi

fig = plt.figure()
p = plt.subplot(111,projection='3d')
data = range(11)
x = np.array(data)

p.plot_wireframe(x,x,x.reshape(1,-1),color='r',linestyle= ':')
p.set_xlabel('sumbu X')
p.set_ylabel('sumbu Y')
p.set_zlabel('sumbu Z')
plt.show()
```
