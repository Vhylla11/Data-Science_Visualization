```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as im
import mpl_toolkits.mplot3d.axes3d
```

```python
fig = plt.figure()
p = plt.subplot(111,projection='3d')

# data
x = np.arange(10)
y = np.arange(10)
z = np.arange(10)

plt.scatter(x,y,z)
plt.show()
```

```python
# untuk buat 3d harus ada 6 data
#maka tambahkan 3 data ( satuan )

fig = plt.figure()
p = plt.subplot(111,projection='3d')

# data
x = np.arange(10)
y = np.arange(10)
z = np.zeros(10)
dx = np.ones(10)
dy = np.ones(10)
dz = np.arange(10)

p.bar3d(x,y,z,dx,dy,dz,
        color=[
            'r','b','pink','brown','y','purple','lightgreen','black','g','k'
        ]
)
p.set_xlabel('Sumbu X')
p.set_ylabel('Sumbu Y')
p.set_zlabel('Sumbu Z')
plt.show()
```

# SEABORN

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as im
import mpl_toolkits.mplot3d.axes3d
import seaborn as sb#pip install seaborn
```

```python
## buat dataframe dulu
df = pd.DataFrame({
    'x': np.arange(10),
    'y':np.arange(10),
    'z' :['Pria','Wanita']*5
})
```

```python
# cek style plt.style.use()
plt.style.use('seaborn')
plt.plot(df.x,df.y)
```

# Buat Plot Pakai SEABORN : Style

```python
sb.set(style='whitegrid')
sb.relplot(data=df,x='x',y='y',kind='line')

## lihat style semua # plt.style.avaliable
```

```python
## atur warna di setiap titik
sb.set(style='whitegrid')
sb.relplot(data=df,x='x',y='y',hue='z',style='z',size='z')
```

```python
sb.relplot(data=df,x='x',y='y',hue='z',style='z',size='z',kind='line')

sb.barplot(data=df,x='x',y='y') ## kalau tanpa relplot maka tambahkan hue di barplot
plt.show()
```

```python
## panggil dataset
df=sb.load_dataset('flights')
## cek dulu ada data kosng atau tdk

sb.barplot(data=df,x='year',y='passengers')
plt.show()
```

```python
# tampilkan data passengers per bulan matlop
df=sb.load_dataset('flights')
df = df.pivot('month','year','passengers')
df = df.sum()
df[1949] ## jumlah pass thun 1949


```

```python
## buat bar pakai index

df=sb.load_dataset('flights')
df = df.pivot('month','year','passengers')
df = df.sum()
df = df.to_frame()
df

sb.barplot(data=df,x=df.index,y=0,orient='v')
plt.show()
```

```python
## panggil heatmap pakai seaborn

df=sb.load_dataset('flights')
df = df.pivot('month','year','passengers')
sb.heatmap(df,annot=True,fmt='d')

## kalau terpotong pakai ylim
# plt.ylim(-.1,2.5)
plt.show()
```

```python
# korelasi

df=sb.load_dataset('flights')
#df = df.pivot('month','year','passengers')
df=df.corr()
df
sb.heatmap(df,annot=True,cmap='hot_r')
```

# FOLIUM

```python
import folium
```

```python
map = folium.Map(
    location =[-6.1753924,106.8271528],tiles='OpenStreetMap',zoom_start = 17
)
## tambahin marker,ada tulisan , dan klu klik muncul tulisan,icon
folium.Marker(location =[-6.1753924,106.8271528],
              tooltip='<img width="100px" src="https://www.pngfans.com/middle-759e80a3ebe9d3e6-cartoon-spongebob-squarepants.html">''<i>Halo!</i>',
              popup='<b>MONAS</b><br><small>Jakarta</small>' ,
              icon= folium.Icon(color='blue',icon='cloud')
             
).add_to(map)
   
    
## tambah marker 
folium.Marker(location =[-6.1752,106.8271],
              tooltip='<i>Halo!</i>',
              popup='<b>MONAS</b><br><small>Jakarta</small>' ,
              icon= folium.Icon(color='red',prefix='fa',icon='save')
             
).add_to(map)

folium.Circle (location =[-6.1752,106.8271],
               radius=40,
               color='red',fill=True,fill_color='yellow'
).add_to (map)
map
```
