# Diagram Batang

```python
df = pd.DataFrame({
'x':[1,2,3,4,5],
'y':[9,5,7,3,6],
'z':[2,3,4,5,3] })

```

```python
plt.bar(df.x,df.y)
plt.title('BARCHAT',fontdict={'family':'impact','size':30,'color':'blue'})
plt.yticks(np.arange(0,12,1))
plt.xlabel ('Kemiskinan',fontdict={'family':'impact','size':20,'color':'red'})
plt.ylabel('penduduk',fontdict={'family':'impact','size':20 ,'color':'red'})
plt.show()
```

```python
## warna bar yang diganti
plt.bar(df.x,df.y,color='pink',edgecolor='black',linewidth=5)

# warna setiap bar berbeda
plt.bar(df.x,df.y,color =['r','g','b','y','w'],edgecolor='black',linewidth=5)
plt.yticks(np.arange(0,12,1))
plt.title('BARCHAT',fontdict={'family':'impact','size':30,'color':'blue'})
plt.xlabel ('Kemiskinan',fontdict={'family':'impact','size':20,'color':'red'})
plt.ylabel('penduduk',fontdict={'family':'impact','size':20 ,'color':'red'})
plt.show()
```

```python
## untuk bar bentuk horizontal
df = pd.DataFrame({
'x':[1,2,3,4,5],
'y':[9,5,7,3,6],
'z':[2,3,4,5,3] })

plt.barh(df.x,df.y,color =['r','g','b','y','w'],edgecolor='black',linewidth=5)
```

```python
## tambahkan ketinggian bar 
plt.bar(df.x,df.y)
plt.title('BARCHAT',fontdict={'family':'impact','size':30,'color':'blue'})
plt.yticks(np.arange(0,12,1))
plt.xlabel ('Kemiskinan',fontdict={'family':'impact','size':20,'color':'red'})
plt.ylabel('penduduk',fontdict={'family':'impact','size':20 ,'color':'red'})
plt.show()
```

```python
## Tampilkan values atau nilai diatas batang
plt.bar(df.x,df.y,color='brown')
plt.yticks(np.arange(0,12,1))
plt.title('BARCHAT',fontdict={'family':'impact','size':30,'color':'blue'})
for i in df['y']:
    plt.text(df[df['y']==i]['x']-.1,i+.2,i,fontdict={'size':20})
plt.xlabel ('Kemiskinan',fontdict={'family':'impact','size':20,'color':'red'})
plt.ylabel('penduduk',fontdict={'family':'impact','size':20 ,'color':'red'})
plt.show()
```

```python
## Tampilkan diagram batang 2 warna
plt.bar(df.x,df.z,color='black',edgecolor='black')
plt.bar(df.x,df.y,color='pink',edgecolor='pink',bottom=df.z,yerr=.5)
plt.yticks(np.arange(0,12,1))
plt.title('BARCHAT',fontdict={'family':'impact','size':30,'color':'blue'})
for i in df['y']:
    plt.text(df[df['y']==i]['x']-.1,i+.2,i,fontdict={'size':20})
plt.xlabel ('Kemiskinan',fontdict={'family':'impact','size':20,'color':'red'})
plt.ylabel('penduduk',fontdict={'family':'impact','size':20 ,'color':'red'})
plt.show()
```

```python
## scatter
plt.plot(df.x,df.y,'Blue',markersize=20)
plt.scatter(df.x,df.y,marker='$*$',s=300,color='black',edgecolor='r')
```

```python
## scatter
plt.plot(df.x,df.y,'r',markersize=20,zorder=1)
plt.scatter(df.x,df.y,marker='$*$',s=300,color='b',edgecolor='r',zorder=2)
plt.bar(df.x,df.y,color='green',zorder=0)
plt.show()
```
