```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
persentase = [35,65]
keterangan = ['A','B']
explode = (0,0.3)
plt.pie ( persentase,labels=keterangan,colors=['b','y'],
        startangle = 90,shadow=True,counterclock=False,
        textprops={'size':18,'color':'pink'},autopct='%.1f%%',explode=explode)

plt.show()
```

```python
## untuk 3 bagian
persentase = [20,45,35]
keterangan = ['A','B','C']
explode = (0,0.3,0)
plt.pie ( persentase,labels=keterangan,colors=['b','y','r'],
        startangle = 90,shadow=True,counterclock=False,
        textprops={'size':18,'color':'pink'},autopct='%1.0f%%',radius=1,explode=explode)

plt.show()
```
